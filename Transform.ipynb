Rename columns

```python
# Rename the column Game.1 as Game
games_df1.rename(columns = {"Game.1": "Game"},inplace = True)
```

Add Release Year Column

```python
# Defining the function for extracting year from Release Date
def year(row):
    yr= int(row.strip().split(' ')[2]) # Taking year apart from the Release Date
    if yr<=22: # Converting extracted year into decade
        return yr+2000
    else:
        return yr+1900
    return yr
```

```python
games_df1 = games_df1.loc[games_df1['Release Date'].notnull()]
```

```python
# Applying the above function on Release Date column and assinging it to new column name Release Year
games_df1["Release Year"]=games_df1.apply(lambda row: year(row['Release Date']),axis=1).copy()
```

```python
games_df1
```

Convert Release date values to Date format (yyyy-mm-dd)

```python
# Defining the function for formating Release Date into YYYY-MM-DD format
def dateformat(row):
    date_list = row.strip().split(' ') # Converting the date into list 
    date_list[0]=date_list[0][0:2] # Removing the suffix i.e. nd,st,rd,th from Date 
    new_date=pd.to_datetime("-".join(date_list)) # Now join the the updated list and make it datetime format
    formated_date = new_date.strftime("%Y-%m-%d") # Changing the datetime format
    return formated_date
```

```python
# Updating the Release Date column values by applying above function
games_df1['Release Date']=games_df1.apply(lambda row: dateformat(row['Release Date']), axis=1)
```

```python
games_df1.head(5)
```

Convert Sales values to Float format

```python
# Creating function to strip out character m from various sales columns 
def to_float(row):
    if type(row)==str: 
        float_value=float(row.replace(row[-1],'')) # Replacing m with empty value and converting it into float
    else:
        float_value=float(row)
    return float_value
```

```python
# Apply 0 to Nan values
games_df1.fillna(value='0', inplace=True)
```

```python
games_df1
```

```python
# Apply above created function on Total Sales 
games_df1['Total Sales']=games_df1.apply(lambda row: to_float(row['Total Sales']), axis=1)
```

```python
# Apply above created function on NA Sales 
games_df1['NA Sales']=games_df1.apply(lambda row: to_float(row['NA Sales']), axis=1)
```

```python
# Apply above created function on PAL Sales 
games_df1['PAL Sales']=games_df1.apply(lambda row: to_float(row['PAL Sales']), axis=1)
```

```python
# Apply above created function on Japan Sales 
games_df1['Japan Sales']=games_df1.apply(lambda row: to_float(row['Japan Sales']), axis=1)
```

```python
# Apply above created function on Other Sales 
games_df1['Other Sales']=games_df1.apply(lambda row: to_float(row['Other Sales']), axis=1)
```

```python
# Apply above created function on Other Sales 
games_df1['Total Shipped']=games_df1.apply(lambda row: to_float(row['Total Shipped']), axis=1)
```
